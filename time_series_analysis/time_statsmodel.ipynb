## Hodrick-prescott filter
### y_t = T_t + C_t
#### time series = trend_component + cyclical_component
#### Lamda value is 1600 for quarterly data, 6.25 for annual data and 129600 for monthly data

```python
df = pd.read_csv("UDEMY_TSA_FINAL/Data/macrodata.csv",index_col=0,parse_dates=True)
```

```python
df.head()
```

```python
df['realgdp'].plot(figsize=(15,5))
```

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
gdp_cycle,gdp_trend = hpfilter(df['realgdp'],lamb=1600)
```

```python
type(gdp_cycle)
```

```python
df['trend']=gdp_trend.plot()
```

```python
df['trend']=gdp_trend
```
